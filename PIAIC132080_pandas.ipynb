```python
import glob

path =r'E:\Panda_Assignment\Assignment'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))
us_census= pd.concat(dfs, ignore_index=True)
us_census
                    
```

```python
us_census.dtypes
```

```python
us_census.drop(columns='Unnamed: 0',inplace=True)
```

```python
us_census.dtypes
```

```python
us_census.head()
```

```python
us_census[['Men','Women']]=us_census.GenderPop.str.split("_",expand=True)
```

```python
us_census.head()
```

```python
us_census.drop(columns='GenderPop',inplace=True)
```

```python
us_census.head()
```

```python
us_census=us_census.replace('M','',regex=True)
us_census=us_census.replace('F','',regex=True)
us_census=us_census.replace('%','',regex=True)
us_census=us_census.replace(' ','',regex=True)
```

```python
us_census.head()
```

```python
us_census['income']=us_census.Income.str.strip('$')
```

```python
us_census.head()
```

```python
us_census.drop(columns='Income',inplace=True)
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census["Hispanic"] = us_census.Hispanic.astype(float)
us_census["White"] = us_census.White.astype(float)
us_census["Black"] = us_census.Black.astype(float)
us_census["Native"] = us_census.Native.astype(float)
us_census["Asian"] = us_census.Asian.astype(float)
us_census["Men"] = us_census.Men.astype(int)
#us_census["Women"] = us_census.Women.astype(float)
us_census["income"] = us_census.income.astype(float)
us_census["Pacific"] = us_census.Pacific.astype(float)
```

```python
us_census.dtypes
```

```python
us_census["Women"]=us_census["TotalPop"]-us_census["Men"]
```

```python
us_census["Women"] = us_census.Women.astype(int)
```

```python
us_census.dtypes
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census['Women'],us_census['income'])
plt.xlabel('Female Population')
plt.ylabel('Income ($)')
plt.show()
```
